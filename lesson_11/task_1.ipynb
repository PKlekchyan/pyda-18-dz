```python
import pandas as pd

data1 = pd.read_csv("movies.csv", encoding="utf-8")

# Получаем DataFrame таблицы с рейтингом, группируем по Id фильма и находим среднее значение рейтинга.
data2 = pd.read_csv("ratings.csv", encoding="utf-8").groupby('movieId').agg('mean').rating.reset_index()

# После этого группируем два DataFrame методом merge, выбирая нужные столбцы
all_data = pd.merge(data1[['movieId', 'title']], data2, on=('movieId'))

def class_movie(row):
    if row['rating'] <=2:
        return "низкий рейтинг"
    elif row['rating'] <= 4:
        return "средний рейтинг"
    elif 4.5 <= row['rating'] <=5:
        return "высокий рейтинг"

# Затем, добавляем в сгруппированный Df новый столбец с указанием рейтинга
all_data['class'] = all_data.apply(class_movie, axis=1)
all_data
```
