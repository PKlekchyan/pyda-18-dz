```python
import pandas as pd
import re

movies = pd.read_csv("movies.csv", encoding="utf-8")
ratings = pd.read_csv("ratings.csv", encoding="utf-8").groupby('movieId').agg('mean').rating.reset_index()
all_data = pd.merge(movies[['movieId', 'title']], ratings, on=('movieId'))

year = [str(i) for i in range(1950, 2011)]

def production_year(param, year):
    pattern = r'(?<=\D)(\d{4})(?=\D)'
    if re.search(pattern, param):
        current_year = re.search(pattern, param)[0]
        if current_year in year:
            return int(current_year)
    return 1900

all_data['year'] = all_data.title.apply(production_year, year=year)
all_data.groupby("year").agg('mean').rating.reset_index().sort_values('year', ascending=False)
```
